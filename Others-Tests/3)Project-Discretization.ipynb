```python
dom_values=np.linspace(-domain,domain,dna_length)
```

```python
function1= np.random.uniform(codomain_min, codomain_max,dna_length)
```

```python
function=np.convolve(function1, np.ones(3)/3, mode='same')
```

```python
plt.plot(dom_values, function)
plt.show()
```

```python
plt.step(dom_values, function, where='post')
```

# FUNZIONI FINALI

Generare funzioni di attivazione:
- Dato un intervallo di dominio [dom_min,dom_max], la lunghezza del [DNA], quanti punti avere in tale intervallo, si generano dei valori di codominio nell'intervallo [v_min,v_max] che poin vengono centrati nella finestra del dominio
- Interpolando i valori si genera una funzione di attivazione

```python
def DNA_generator(dna_length,v_min,v_max):
    # Method 1
    DNA=np.random.uniform(v_min, v_max, dna_length) 
    # 
    # # Method 2 
    # DNA=np.random.randint(v_min,v_max,dna_length)
    
    # # Method 3     
    # points=np.linspace(v_min,v_max,dna_length)
    # DNA=np.random.choice(points,dna_length,replace=False))
    return DNA
    
    
def act_func_generator(DNA,dom_min,dom_max):
    x_dom=np.linspace(dom_min,dom_max,len(DNA))
    function=interp1d(x_dom,DNA,kind='cubic')
    return function
    
```

```python
length_DNA=10
v_min=0
v_max=2
A=3
levels=DNA_generator(length_DNA,v_min,v_max)
function=act_func_generator(levels,-A,A)
```

```python
x_doms=np.linspace(-A,A,1000)
```

```python
y_doms=function(x_doms)
```

```python
plt.plot(x_doms,y_doms)
```
